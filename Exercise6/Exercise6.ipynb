```python
import pandas as pd
import plotly.graph_objects as go
```

```python
arabica_data = pd.read_csv('../data/arabica_data_cleaned.csv')
arabica_df = pd.DataFrame(arabica_data)
arabica_df
```

```python
arabica_df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
arabica_df = arabica_df[["Country.of.Origin", "Producer", "Processing.Method"]]
arabica_df
```

```python
arabica_df.sort_values(['Country of Origin'])
```

```python
for column in ["Country of Origin", "Producer", "Processing Method"]:
    graph = [
        go.Bar(
            x=arabica_df[column].unique(),
            y=arabica_df[column].value_counts()
        )
    ]
    fig = go.Figure(data=graph)
    fig.update_layout(title=('Arabica '+ column), yaxis=dict(title='counts'), xaxis=dict(title=column))
    fig.show()
```
